```python
def plot_silhouette(sscore, ypreds, n):
    for ypred in ypreds:
        clusters = set(ypred)
        if len(clusters) != n: continue
        plt.figure(figsize=(4,len(clusters)*2))
        for i in clusters:
            plt.subplot(len(clusters),1,i+1)
            plt.hist(silScore[6][ypred==i], orientation='horizontal')
            plt.xlim((0,8))
            plt.xticks([])
        plt.tight_layout()

clusters = np.arange(2,10)
silScore = []
ypreds = []

X = Xbreast
for n in clusters:
    km = KMeans(n_clusters=n)
    km.fit(X)
    ypred = km.predict(X)
    silScore.append(silhouette_samples(X, ypred))
    ypreds.append(ypred)
    
plot_silhouette(silScore, ypreds, n=2)
```

### Expectation Maximization

```python
from collections import defaultdict

clusters = np.arange(1,10)
covTypeList = ['full', 'tied', 'diag', 'spherical']

for X in [Xiris, Xbreast]:
    aicDict = defaultdict(list)
    bicDict = defaultdict(list)
    for covType in covTypeList:
        for n in clusters:
            gm = GaussianMixture(n_components=n, covariance_type=covType, max_iter=500, n_init=5, reg_covar=1e-2)
            gm.fit(X)
            aic = gm.aic(X)
            bic = gm.bic(X)
            aicDict[covType].append(aic)
            bicDict[covType].append(bic)

    plt.figure(figsize=(12,4))


    plt.subplot(1,2,1)
    for key, value in aicDict.items():
        plt.plot(clusters, value, label=key)
    plt.ylabel('AIC score')
    plt.xlabel("number of clusters")
    plt.yscale('log')
    plt.legend()
    plt.title('AIC')

    plt.subplot(1,2,2)
    for key, value in bicDict.items():
        plt.plot(clusters, value, label=key)
    plt.ylabel('BIC score')
    plt.xlabel("number of clusters")
    plt.yscale('log')
    plt.legend()
    plt.title('BIC')

    plt.tight_layout()
```

```python
# confusion matrix
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score

gm = GaussianMixture(n_components=3, covariance_type='full', max_iter=500, n_init=5, reg_covar=1e-2)
gm.fit(Xiris)
ypred = gm.predict(Xiris)

cm = confusion_matrix(yiris, ypred)
sns.heatmap(cm, annot=True, cmap="YlGnBu")
plt.ylabel('actual')
plt.xlabel('prediction')
print(accuracy_score(yiris, ypred))
```

### Random Projection

```python
X = Xiris
irisComponents = np.arange(1,5)
irisLossList = []

for n in irisComponents:
    rp = GaussianRandomProjection(n_components=n, eps=0.1, random_state=5)
    rp.fit(X)
    Xproj = rp.transform(X)

    Xrecon = np.matmul(rp.components_.T, Xproj.T).T
    loss = ((X-Xrecon)**2).mean()
    irisLossList.append(loss)
    
#----------------------------
X = Xbreast
breastComponents = np.arange(1,11)
breastLossList = []

for n in breastComponents:
    rp = GaussianRandomProjection(n_components=n, eps=0.1, random_state=5)
    rp.fit(X)
    Xproj = rp.transform(X)

    Xrecon = np.matmul(rp.components_.T, Xproj.T).T
    loss = ((X-Xrecon)**2).mean()
    breastLossList.append(loss)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(irisComponents, irisLossList)
plt.title("Iris")
plt.xlabel('number of component')
plt.ylabel('information loss')

plt.subplot(1,2,2)
plt.plot(breastComponents, breastLossList)
plt.title("Breast Cancer")
plt.xlabel('number of component')
plt.ylabel('information loss')

plt.tight_layout()
```

## Dimensionality Reduction and Clustering

```python
X = Xiris
y = yiris

algs = {'PCA': PCA(n_components=2),
        'ICA': FastICA(n_components=3),
        'RP': GaussianRandomProjection(n_components=3, random_state=5),
        'RF': RandomForestClassifier()}

for k in algs.keys():
    if k == 'RF':
        model = algs[k]
        rf = model.fit(X, y)
        ascending_order = np.argsort(rf.feature_importances_)
        feature_index = ascending_order[::-1][:3]
        Xproj = X[:,feature_index]
        
    else:
        model = algs[k]
        Xproj = model.fit_transform(X)
        
    km = KMeans(n_clusters=3)
    km.fit(Xproj)
    cm = confusion_matrix(y, km.labels_)
    sns.heatmap(cm, annot=True, cmap="YlGnBu")
    plt.ylabel('actual')
    plt.xlabel('prediction')
    plt.title(k)
    plt.show()
```

## Neural Net

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
from keras.layers import Dropout

from sklearn.metrics import recall_score, precision_score, accuracy_score

def create_neural_net(category=4, dim=40):
    model = Sequential()
    model.add(Dense(1024, input_dim=dim, activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(512, activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(128, activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(category, activation="sigmoid"))
    return model

def plot_DNN(model):
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(model.history.history['acc'], label="training accuracy")
    plt.plot(model.history.history['val_acc'], label="testing accuracy")
    plt.legend(loc="best")
    plt.xlabel("Epoch")
    plt.ylabel("accuracy")

    plt.subplot(1,2,2)
    plt.plot(model.history.history['loss'], label="training loss")
    plt.plot(model.history.history['val_loss'], label="testing loss")
    plt.legend(loc="best")
    plt.xlabel("Epoch")
    plt.ylabel("loss function")
    plt.tight_layout()
    plt.show()
    
    return None
```

### PCA 

```python
Xpca = PCA(n_components=3).fit_transform(Xotto)

XPcaTrain, XPcaTest, yPcaTrain, yPcaTest \
= train_test_split(Xpca, yotto, test_size=0.2)

dnnPca = create_neural_net(category=4, dim=XPcaTrain.shape[1])
callbacks = [EarlyStopping(monitor="loss", patience=10, mode="min")]
dnnPca.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
dnnPca.fit(XPcaTrain, 
             yPcaTrain, 
             epochs=2000, 
             batch_size=10000, 
             validation_data=(XPcaTest, yPcaTest),
             callbacks=callbacks)
    
plot_DNN(dnnPca)
```

```python
ypred = dnnPca.predict(XPcaTest)

yPcaPred = np.zeros(ypred.shape[0])

for i, p in enumerate(ypred):
    yPcaPred[i] = np.argmax(p)
cm = confusion_matrix(yPcaTest, yPcaPred)
sns.heatmap(cm, annot=True, fmt='d', cmap="YlGnBu")
plt.ylabel('actual')
plt.xlabel('prediction')
plt.title("PCA")
plt.show()

accuracy_score(yPcaTest, yPcaPred)
```

```python
from collections import defaultdict

clusters = np.arange(1,19)
covTypeList = ['full', 'tied', 'diag', 'spherical']

aicDict = defaultdict(list)
bicDict = defaultdict(list)
for covType in covTypeList:
    for n in clusters:
        gm = GaussianMixture(n_components=n, covariance_type=covType) # , max_iter=500, n_init=5, reg_covar=1e-2)
        gm.fit(Xotto)
        aic = gm.aic(Xotto)
        bic = gm.bic(Xotto)
        aicDict[covType].append(aic)
        bicDict[covType].append(bic)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
for key, value in aicDict.items():
    plt.plot(clusters, value, label=key)
plt.ylabel('AIC score')
plt.xlabel("number of clusters")
# plt.yscale('log')
plt.legend()
plt.title('AIC')

plt.subplot(1,2,2)
for key, value in bicDict.items():
    plt.plot(clusters, value, label=key)
plt.ylabel('BIC score')
plt.xlabel("number of clusters")
# plt.yscale('log')
plt.legend()
plt.title('BIC')

plt.tight_layout()
```
