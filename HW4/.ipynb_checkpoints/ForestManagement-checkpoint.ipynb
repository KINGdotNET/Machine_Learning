```python
import seaborn as sns
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm

from mdptoolbox.mdp import ValueIteration, PolicyIteration
from mdptoolbox.example import forest
```

## Value Iteration

```python
cols = ['R1', 
        'R2', 
        'P', 
        'Number of States', 
        'Number of Iterations', 
        'Total Time']

r1List = [2,4]
r2List = [4,2]
pList = [0.05, 0.1, 0.3]
states = [500, 2000, 5000]
r1 = 4
r2 = 2
p = 0.05

dfVi = []
for r1, r2 in zip(r1List, r2List):
    for p in pList:
        for S in states:
            P, R = forest(S=S, p=p, r1=r1, r2=r2)
            vi = ValueIteration(P, R, 
                                discount=0.99, 
                                max_iter=5000, 
                                epsilon=1e-7)
#             vi.setVerbose()
            vi.run()
            dfVi.append([r1, r2, p, S, len(vi.iterVar), vi.time])
```

```python
dfVi = pd.DataFrame(dfVi, columns=cols)
dfVi
```

## Policy Iteration

```python
cols = ['R1', 
        'R2', 
        'P', 
        'Number of States', 
        'Number of Iterations', 
        'Total Time']

r1List = [2,4]
r2List = [4,2]
pList = [0.05, 0.1, 0.3]
states = [500, 2000, 5000]

dfPi = []
for r1, r2 in tqdm(zip(r1List, r2List)):
    for p in pList:
        for S in states:
            P, R = forest(S=S, p=p, r1=r1, r2=r2)
            pi = PolicyIteration(P, R, 
                                discount=0.99, 
                                max_iter=5000, 
                                eval_type=1)
            pi.run()
            dfPi.append([r1, r2, p, S, len(pi.iterVar), pi.time])
```

```python
dfPi = pd.DataFrame(dfPi, columns=cols)
dfPi
```

```python
r1 = 40
r2 = 2
p = 0.05

P, R = forest(S=1000, p=p, r1=r1, r2=r2)
# R[:,0][1:-1] = 1
# R[:,1][-1] = r2
pi = PolicyIteration(P, R, discount=0.99, max_iter=5000, eval_type=1)
pi.setVerbose()
pi.run()
```
