```python
import seaborn as sns
from gym.envs.toy_text.frozen_lake import FrozenLakeEnv
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
import pprint
pp = pprint.PrettyPrinter(indent=2, width=80)

from mdptoolbox.mdp import ValueIteration, PolicyIteration, QLearning, FiniteHorizon

env4 = FrozenLakeEnv(map_name='4x4', is_slippery=False)
env20 = FrozenLakeEnv(map_name='20x20', is_slippery=False)
env40 = FrozenLakeEnv(map_name='40x40', is_slippery=False)
```

```python
def transform(env):
    P = np.zeros((4,env.nS,env.nS))
    R = np.zeros(env.nS,)

    for s in range(env.nS):
        for a in range(4):
            for p, s_, r, _ in env.P[s][a]:
                P[a][s][s_]+=p
                if s!=s_:
                    R[s_]=r
    return P, R
```

## Sanity Check


```python
env = env4

P, R = transform(env)

vi = ValueIteration(P, R, discount=0.99, max_iter=5000, epsilon=0.001)
vi.setVerbose()
vi.run()
```

```python
sc_policy = np.array(vi.policy).reshape(4,4)
pp.pprint(sc_policy)
```

## Value Iteration
### 20x20 no slip

```python
env20_noslip = FrozenLakeEnv(map_name='15x15', is_slippery=False)

P, R = transform(env20_noslip)
vi20_noslip = ValueIteration(P, R, discount=0.99, max_iter=5000, epsilon=0.001)
vi20_noslip.setVerbose()
vi20_noslip.run()
```

```python
plt.plot(vi20_noslip.iterVar)
plt.yscale('log')
plt.xlabel('iterations')
plt.ylabel('value difference')
```

```python
vi_value = np.array(vi20_noslip.V).reshape(int(np.sqrt(env20_noslip.nS)),int(np.sqrt(env20_noslip.nS)))

vi_value[vi_value==0] = np.nan

sns.heatmap(vi_value)
```

```python
policy = np.array(vi20_noslip.policy).reshape(int(np.sqrt(env20_noslip.nS)),int(np.sqrt(env20_noslip.nS)))

policy = policy.astype(float)
policy[np.isnan(vi_value)] = np.nan

sns.heatmap(policy)
```

### 20x20 slip

```python
env20_slip = FrozenLakeEnv(map_name='20x20', is_slippery=True)

P, R = transform(env20_slip)
vi20_slip = ValueIteration(P, R, discount=0.99, max_iter=5000, epsilon=0.001)
vi20_slip.setVerbose()
vi20_slip.run()
```

```python
plt.plot(vi20_slip.iterVar)
plt.yscale('log')
plt.xlabel('iterations')
plt.ylabel('value difference')
```

```python
value = np.array(vi20_slip.V).reshape(int(np.sqrt(env20_slip.nS)),int(np.sqrt(env20_slip.nS)))

value[value==0] = np.nan

sns.heatmap(value)
```

## Policy Iteration
### 20x20 no slip

```python
env20_noslip = FrozenLakeEnv(map_name='20x20', is_slippery=False)

P, R = transform(env20_noslip)

policy0 = np.zeros(env20_noslip.nS)+2
pi20_noslip = PolicyIteration(P, R, 
                              discount=0.99, 
                              max_iter=5000,
                              eval_type=1,
                             policy0=policy0)
pi20_noslip.setVerbose()
pi20_noslip.run()
```

```python
plt.plot(pi20_noslip.iterVar)
plt.yscale('log')
plt.xlabel('iterations')
plt.ylabel('action difference')
```

```python
pi_value = np.array(pi20_noslip.V).reshape(int(np.sqrt(env20_noslip.nS)),int(np.sqrt(env20_noslip.nS)))

pi_value[pi_value<=2e-4] = np.nan

sns.heatmap(pi_value)
```

### 20x20 slip

```python
env20_slip = FrozenLakeEnv(map_name='20x20', is_slippery=True)


policy0 = np.zeros(env20_slip.nS)+2
P, R = transform(env20_slip)
pi20_slip = PolicyIteration(P, R, 
                            discount=0.99, 
                            max_iter=5000, 
                            eval_type=1,
                           policy0=policy0)
pi20_slip.setVerbose()
pi20_slip.run()
```

```python
plt.plot(pi20_slip.iterVar)
plt.yscale('log')
plt.xlabel('iterations')
plt.ylabel('action difference')
```

```python
pi_value = np.array(pi20_slip.V).reshape(int(np.sqrt(env20_noslip.nS)),int(np.sqrt(env20_noslip.nS)))

pi_value[pi_value<=1e-6] = np.nan

sns.heatmap(pi_value)
```

## VI vs. PI
### no slip

```python
pi20_noslip_policy = \
np.array(pi20_noslip.policy).reshape(int(np.sqrt(env20_noslip.nS)),int(np.sqrt(env20_noslip.nS)))

pi20_noslip_policy = pi20_noslip_policy.astype(np.float)
pi20_noslip_policy[np.isnan(vi_value)] = np.nan

vi20_noslip_policy = \
np.array(vi20_noslip.policy).reshape(int(np.sqrt(env20_noslip.nS)),int(np.sqrt(env20_noslip.nS)))

vi20_noslip_policy = vi20_noslip_policy.astype(np.float)
vi20_noslip_policy[np.isnan(vi_value)] = np.nan

sns.heatmap(vi20_noslip_policy)
plt.show()

sns.heatmap(pi20_noslip_policy)
plt.show()

sns.heatmap(pi20_noslip_policy-vi20_noslip_policy)
plt.show()
```
